i) Representa gráficamente las datos de la columna Height de forma que se aprecie la distribución de los datos.

```r
heights <- nba$Height
hist(heights, 
          breaks = 10, 
          xlim = c(min(heights, na.rm = TRUE) - 5,max(heights, na.rm = TRUE) + 10),
          xlab = "Alturas, cm",
          ylab = "Frecuencia",
          main="Histograma de alturas")
```

ii) Calcula la media muestral y represéntala mediante una línea en la gráfica.

```r
height_m <- mean(heights, na.rm = TRUE)
print(height_m)
```

```r
hist(heights, 
          breaks = 10, 
          xlim = c(min(heights, na.rm = TRUE) - 5,max(heights, na.rm = TRUE) + 10),
          xlab = "Alturas, cm",
          ylab = "Frecuencia",
          main="Histograma de alturas")
abline(v = height_m, col = "royalblue", lwd = 2)
legend(x = "topright", # location of legend within plot area
 c("Media"),
 col = c("royalblue"),
 lwd = c(1, 2))
```

iii)  Calcula la desviación típica muestral y representa el intervalo media muestral ± desviación típica muestral en la gráfica.

```r
#height_med <- median(heights, na.rm = TRUE)
#print(height_med)
height_sd <- sd(heights, na.rm = TRUE)
print(height_sd)
```

```r
hist(heights, 
          breaks = 10, 
          xlim = c(min(heights, na.rm = TRUE) - 5,max(heights, na.rm = TRUE) + 10),
          xlab = "Alturas, cm",
          ylab = "Frecuencia",
          main="Histograma de alturas")
abline(v = height_m, col = "royalblue", lwd = 2)
abline(v = height_m - height_sd, col = "darkgreen", lwd = 2)
abline(v = height_m + height_sd, col = "darkgreen", lwd = 2)
legend(x = "topright", # location of legend within plot area
 c("Media","Media±desv. típ."),
 col = c("royalblue", "darkgreen"),
 lwd = c(2, 2))
```

```r
h <- hist(heights, 
          breaks = 10, 
          #xlim = c(min(heights, na.rm = TRUE) - 5,max(heights, na.rm = TRUE) + 10),
          #xlim = c(170,230),
          ylim = c(0,110),
          xlab = "Alturas, cm",
          ylab = "Fecuencia",
          main="Histograma con fit normal")
xfit<-seq(min(heights, na.rm = TRUE),max(heights, na.rm = TRUE),length=40)
yfit<-dnorm(xfit,mean=mean(heights, na.rm = TRUE),sd=sd(heights, na.rm = TRUE))
yfit <- yfit*diff(h$mids[1:2])*length(heights)
lines(xfit, yfit, col="red", lwd=2)
abline(v = height_m, col = "royalblue", lwd = 2)
legend(x = "topright", # location of legend within plot area
 c("Normal fit", "Media"),
 col = c("red", "royalblue"),
 lwd = c(2, 2))
```
