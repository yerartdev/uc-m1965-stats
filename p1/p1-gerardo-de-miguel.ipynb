```r
heights <- nba$Height
#tallest <- max(heights, na.rm = TRUE)
#print(tallest)
height_m <- mean(heights, na.rm = TRUE)
print(height_m)
height_med <- median(heights, na.rm = TRUE)
print(height_med)
height_sd <- sd(heights, na.rm = TRUE)
print(height_sd)
```

```r
hist(heights, 
          breaks = 10, 
          prob = TRUE,
          xlim = c(min(heights, na.rm = TRUE) - 5,max(heights, na.rm = TRUE) + 10),
          #xlim = c(170,230),
          ylim = c(0,0.05),
          xlab = "Alturas, cm",
          ylab = "Densidad",
          main="Histograma con densidad, media y desviacin tpica")
abline(v = height_m, col = "royalblue", lwd = 2)
lines(density(heights, na.rm = TRUE), # density plot
 lwd = 2, # thickness of line
 col = "chocolate3")
#md <- abline(v = median(heights, na.rm = TRUE), col = "red",  lwd = 2)
legend(x = "topright", # location of legend within plot area
 c("Density plot", "Media"),
 col = c("chocolate3", "royalblue"),
 lwd = c(2, 2))
```

```r
h <- hist(heights, 
          breaks = 10, 
          #xlim = c(min(heights, na.rm = TRUE) - 5,max(heights, na.rm = TRUE) + 10),
          #xlim = c(170,230),
          ylim = c(0,110),
          xlab = "Alturas, cm",
          ylab = "Fecuencia",
          main="Histograma con fit normal")
xfit<-seq(min(heights, na.rm = TRUE),max(heights, na.rm = TRUE),length=40)
yfit<-dnorm(xfit,mean=mean(heights, na.rm = TRUE),sd=sd(heights, na.rm = TRUE))
yfit <- yfit*diff(h$mids[1:2])*length(heights)
lines(xfit, yfit, col="red", lwd=2)
abline(v = height_m, col = "royalblue", lwd = 2)
legend(x = "topright", # location of legend within plot area
 c("Normal fit", "Media"),
 col = c("red", "royalblue"),
 lwd = c(2, 2))
```
