Consideremos un sistema en el que existe una característica x y una variable dependiente y que se relacionan como y = a * x. Supongamos también que para cada valor fijo de x la medida y viene dada por una pdf gaussiana centrada en a * x y con una sigma. 

**Escribe una funcion de R** que tome como input un vector aleatorio de N elementos de x que tome valores entre [0, 10], el factor a, y sigma; y genere el correspondiente vector y.

::GMG::Ayuda:

 - (y - ax) = gauss(ax, sigma) --> y(i) = ax(i) + normal (0, sigma) --> simular un proceso de medida real
 - a, sigma y un vector x  son los argumentos de la función

```r
medida <- function(x,a=1,sigma=0.1) {
    return (a*x + rnorm(length(x), 0, sigma))
}
```

Una vez terminada la función genera el vector x con N = 100, y un vector y con a = 2 y sigma = 0.2. Pintalos en una grafica.  (x,y de puntos)

```r
#::GMG::Referencias
#       https://stackoverflow.com/questions/28376766/r-creating-a-vector-with-a-specific-amount-of-random-numbers
x <- runif(n = 100, min = 0, max = 10)
a <- 2
sigma <- 0.2
y <- medida(x = x,a = a,sigma = sigma)
plot(x,y, type = "p",
     main = "Medida simulada con errores",
     sub = " N(0, sigma) ~ Gauss(ax,sigma)")
```
