1.- Generar una muestra de tamaño N = 10000 correspondiente a la altura de personas adultas, asumiendo que su densidad de probabilidad es una función normal/gaussiana con μ= 1.78m y σ= 0.1 m. Dibuja la densidad de frecuencia y la densidad de probabilidad por separado. Compara μ y σ con la media muestral y la varianza muestral. 

```r
#::GMG::https://www.statmethods.net/graphs/density.html
#       https://www.statmethods.net/advgraphs/probability.html
#       https://www.rdocumentation.org/packages/graphics/versions/3.5.1/topics/hist
#       http://www.sthda.com/english/articles/32-r-graphics-essentials/133-plot-one-variable-frequency-graph-density-distribution-and-more/#one-continuous-variable
#       https://stats.stackexchange.com/questions/52754/is-it-appropriate-to-plot-the-mean-in-a-histogram/52758#52758
#       http://www.r-tutor.com/elementary-statistics
#       https://onlinecourses.science.psu.edu/stat414/node/167/
mu <- 1.78
sigma <- 0.1
x <- rnorm(n = 10000,mean = mu,sd = sigma)
hist(x = x,freq = FALSE, breaks = "Sturges",main = 'Histogram of sample x')
specify_decimal <- function(x, k) trimws(format(round(x, k), nsmall=k))
paste('sample mean:', specify_decimal(mean(x),5),
      'sample variance:',specify_decimal(var(x),5),
      'sample std dev.:',specify_decimal(sd(x),5))
paste('distribution mean:',mu,
      'distribution variance:',sigma**2,
      'distribution std. dev.:',sigma)
#plot(density(x = x),main = 'Density N(μ= 1.78m, σ= 0.1 m)')
```

```r
h<-hist(x = x,breaks = "Sturges")
xfit<-seq(min(x),max(x),length=40) 
yfit<-dnorm(xfit,mean=mean(x),sd=sd(x)) 
yfit <- yfit*diff(h$mids[1:2])*length(x) 
lines(xfit, yfit, col="blue", lwd=2)
```
